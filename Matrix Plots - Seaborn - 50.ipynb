```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True) # Annot = true will show numerical values
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm') 
```

```python
fp= flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,linecolor='black',linewidth=3)
```

```python
# Cluster map use hierarchial clustering for producing a clustered version of heat map

sns.clustermap(fp)
```

```python
sns.clustermap(fp,standard_scale=1)# It will show scale from 0 to 1
```
