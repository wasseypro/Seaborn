```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
# Bar Plot is a basic categorical plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std) # Estimator shows calculations
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.max) # Estimator shows calculations
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.min)
```

```python
# Count plot is same as bar plot except the estimator is explicitly counting the number of occurences

sns.countplot(x='sex',data=tips)
```

```python
# Box plots and violin plots are used to show the distribution of categorical data a box plot is also
# sometimes known as box or whisker plot It will take categorical and numerical data for comparisons

sns.boxplot(x='day', y='total_bill',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill',data=tips)
```

```python
sns.violinplot(x='day', y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)

# Donot use for large dataset
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar',hue='sex')
```
